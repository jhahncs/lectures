```python
import csv

csvfile = open('/content/웰니스_대화_스크립트_데이터셋.csv', newline='', encoding="utf-8-sig")
line_list = list(csv.DictReader(csvfile))
lines = 0
for row in line_list:
  print(row['구분']," | ",row['유저']," | ",row['챗봇'])
  lines += 1
  if lines >= 5:
    break
```

```python
class_set = set()
for row in line_list:
  class_set.add(row['구분'])
print(class_set)
print(len(class_set))
```

```python
class_set = set()
for row in line_list:
  class_set.add(row['구분'].split("/")[0])
print(class_set)
print(len(class_set))
```

```python
for row in line_list:
  row['구분'] = row['구분'].split("/")[0]

with open('output.csv', 'w', encoding='utf-8-sig', newline='') as output_file:
  fc = csv.DictWriter(output_file,fieldnames=line_list[0].keys())
  fc.writeheader()
  fc.writerows(line_list)
```

```python

class_freq = {} # dictionary := a set of key-value pairs
for row in line_list:
  if row['구분'] not in class_freq:
    class_freq[row['구분']] = 0
  class_freq[row['구분']] += 1
  
#print(class_freq)
print(class_freq.keys())
print(class_freq.values())
plt.bar(class_freq.keys(), class_freq.values())
plt.xticks(rotation=90)
plt.show()
```

```python
hannanum.nouns("더 이상 내 감정을 내가 컨트롤 못 하겠어")
```

```python
lines = 0
for row in line_list:
  nouns = hannanum.nouns(row['유저'])
  print(row['구분'],"/",row['유저'],"/",nouns)
  lines += 1
  if lines >= 5:
    break
```

```python
term_list = {}
for row in line_list:
  nouns = hannanum.nouns(row['유저'])

  if row['구분'] not in term_list:
    term_list[row['구분']] = []
    
  for noun in nouns:
    term_list[row['구분']].append(noun)

print(term_list)
```

```python
with open('/content/stopwords.txt') as f:
  stopwords = f.read().splitlines()

print(stopwords)
```

```python


term_list = {}
for row in line_list:
  nouns = hannanum.nouns(row['유저'])

  if row['구분'] not in term_list:
    term_list[row['구분']] = []

  for noun in nouns:
    if noun in stopwords:
      continue
    term_list[row['구분']].append(noun)

print(term_list)
```

```python
from collections import Counter

for key in term_list:
  count = Counter(term_list[key])
  print(key, count)
```

```python
wordcloud = WordCloud(
    font_path = '/content/NanumGothic.ttf',
    width = 800,
    height = 800,
    background_color="white"
)
for key in term_list:
  count = Counter(term_list[key])
  wordcloud = wordcloud.generate_from_frequencies(count)
  
  fig = plt.figure(figsize=(10, 10))
  plt.imshow(wordcloud.to_array(), interpolation="bilinear")
  plt.title(key)
  #plt.show()
  fig.savefig("output/"+key+'.png')

```

```python
import pandas as pd
maxrank = 10
ind = pd.MultiIndex.from_product([term_list.keys(), ["단어","빈도수"]])
print(ind)
df = pd.DataFrame(columns=ind, index=range(1,maxrank+1))
```

```python
for key in term_list:
  count = Counter(term_list[key])
  rank = 1  
  for k,v in count.most_common():
    df.loc[rank,(key, "단어")] = k
    df.loc[rank,(key, "빈도수")] = int(v)    
    rank += 1
    if rank > maxrank:
      break
df
```
