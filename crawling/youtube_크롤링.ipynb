```python
import json
import csv

# ÏáºÏ∏† ÏòÅÏÉÅ Í≤ÄÏÉâ
next_page_token = None
page_index = 0

f = open("output.csv", 'w', encoding='utf-8-sig', newline='')
csv_writer = csv.writer(f, delimiter=',')
csv_writer.writerow(['Page','Title','URL'])
f.close()


while page_index < 2:  # 100Í∞ú Ïù¥ÏÉÅÏùò ÏòÅÏÉÅÏùÑ Í∞ÄÏ†∏Ïò§ÎèÑÎ°ù ÏÑ§Ï†ï

  page_index += 1
  search_response = youtube.search().list(
      part='snippet',
      maxResults=50,  # Í∞ÄÏ†∏Ïò¨ ÏòÅÏÉÅ Ïàò (ÏµúÎåÄ 50Í∞ú)
      q='ÏïÑÍ∏∞ÏÉÅÏñ¥ ÎÖ∏Îûò',  # Í≤ÄÏÉâÏñ¥
      type='video',
      videoDuration='short',  # ÏòÅÏÉÅ Í∏∏Ïù¥Î•º ÏáºÏ∏†Î°ú Ï†úÌïú
      pageToken=next_page_token  # Îã§Ïùå ÌéòÏù¥ÏßÄ ÌÜ†ÌÅ∞ ÏÑ§Ï†ï
  ).execute()
  print(search_response)
  # Í≤ÄÏÉâ Í≤∞Í≥º Ï∂úÎ†•


  f = open("output.csv", 'a', encoding='utf-8-sig', newline='')
  csv_writer = csv.writer(f, delimiter=',')

  for search_result in search_response.get('items', []):
      video_id = search_result['id']['videoId']
      video_title = search_result['snippet']['title']
      video_url = f'https://www.youtube.com/watch?v={video_id}'
            
      onerow = []
      onerow.append(page_index)
      onerow.append(video_title)
      onerow.append(video_url)
      csv_writer.writerow(onerow)
  f.close()

  # Îã§Ïùå ÌéòÏù¥ÏßÄ ÌÜ†ÌÅ∞ ÏÑ§Ï†ï
  next_page_token = search_response.get('nextPageToken')

  if not next_page_token:
      break
```
